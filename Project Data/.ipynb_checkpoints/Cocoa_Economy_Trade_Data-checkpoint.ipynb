## Cocoa Economy Data

```python
# Imports
import panel as pn
pn.extension('plotly')
import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import numpy as np
import os
from pathlib import Path
from dotenv import load_dotenv

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Load data
cocoa_economy_path = Path('/Users', 'tylerhubbard', 'umnfintech', 'project1_group3', 'Resources', 'FAOSTAT_cocoa_trade_matrix.csv')
```

```python
# View a sample of the data for AAPL Historical.
cocoa_econ_df = pd.read_csv(cocoa_economy_path)
cocoa_econ_df = cocoa_econ_df.rename(columns={'Reporter Countries' : 'Area'})
cocoa_econ_df = cocoa_econ_df.set_index(['Element', 'Area', 'Year'])
cocoa_econ_df = cocoa_econ_df.drop(columns=['Domain Code', 'Domain', 'Reporter Country Code (FAO)', 'Partner Country Code (FAO)',
                                            'Partner Countries', 'Element Code', 'Item Code', 'Year Code', 'Flag',
                                            'Flag Description', 'Item'])
cocoa_econ_df = cocoa_econ_df.dropna()
cocoa_econ_df.head(60)
```

```python
export_quantity_df = cocoa_econ_df.drop(index='Export Value')
export_quantity_df = export_quantity_df.rename(columns={'Value' : 'Value (tonnes)'})
export_quantity_df = export_quantity_df.drop(columns='Unit')
export_quantity_df = export_quantity_df.droplevel(level=0)
export_quantity_df
```

```python
export_value_df = cocoa_econ_df.drop(index='Export Quantity')
export_value_df = export_value_df.rename(columns={'Value' : 'Value (1000 US$)'})
export_value_df = export_value_df.drop(columns='Unit')
export_value_df = export_value_df.droplevel(level=0)
export_value_df
```

```python
export_mean = pd.read_csv(cocoa_economy_path, infer_datetime_format=True, parse_dates=True)
export_mean = export_mean.set_index(['Year'])
export_mean = export_mean.rename(columns={'Reporter Countries' : 'Area'})
export_mean = export_mean.drop(columns=['Area', 'Reporter Country Code (FAO)',
                                                          'Partner Country Code (FAO)', 'Element Code',
                                                          'Item Code', 'Year Code'])
export_mean = export_mean.groupby(by=['Element', 'Year'], dropna=False).mean()
export_mean = export_mean.rename(columns={'Value' : 'Value'})
export_mean
```

```python
export_quantity_mean_plot = export_mean.drop(index='Export Value')
export_quantity_mean_plot = export_quantity_mean_plot.plot(kind='bar', figsize=(30,20), rot=45,
                                                           title='Export quantity (tonnes)',
                                                           ylim=(15000, 28000), ylabel='tonnes')
export_quantity_mean_plot
```

```python
export_value_mean_plot = export_mean.drop(index='Export Quantity')
export_value_mean_plot = export_value_mean_plot.plot(kind='bar', figsize=(30,20), rot=45,
                                                           title='Export Value', ylim=(50000, 80000),
                                                    ylabel='Value in 1000$')
export_value_mean_plot
```
